```python
from flask import Flask, render_template, request,jsonify
from flask_cors import CORS,cross_origin
import requests
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen as uReq
```

```python
flipkart_url = "https://www.flipkart.com/search?q=" + "iphone"
```

```python
flipkart_url
```

```python
#1.open the url

uClient = uReq(flipkart_url) # to open a url
```

```python
uClient
```

```python
#2.read the url

flipkartPage = uClient.read() #raw data #right click on the page--view page source

```

```python
flipkartPage
```

```python
uClient.close()
```

```python
flipkart_html = bs(flipkartPage, "html.parser") #beautifying..for readable format,html

```

```python
flipkart_html
```

```python
#<div class="_1AtVbE col-12-12"> 

bigboxes = flipkart_html.findAll("div", {"class": "_1AtVbE col-12-12"})
```

```python
productLink = "https://www.flipkart.com" + box.div.div.div.a['href'] #in class": "_1AtVbE col-12-12 ,after 3 div we got the the href

```

# #scrap review from productlink

```python
prodRes = requests.get(productLink)
prodRes.encoding = 'utf-8'
prod_html = bs(prodRes.text, "html.parser")

print(prod_html)
```

```python
#<div class="_16PBlm">

commentboxes = prod_html.findAll('div', {'class': "_16PBlm"})
```
